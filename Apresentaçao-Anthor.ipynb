## 1.2 Pesquisas Demográficas - População e Renda

Dadas as limitações de acesso a dados de comércio discriminados municipalmente, a melhor forma de tentar atuar com algum tipo de balizamento em dados é buscar aproximações a partir do que é disponível para trabalharmos. Se tivéssemos dados dos municípios sobre número de varejistas, receitas e volumes, certamente teríamos maior confiança na definição de cidades-chave para implementamento das atividades. Como não temos, tentaremos realizar um exercício de abstração, saindo da visão da oferta e partindo para a demanda de produtos finais do varejo. 

Quem são os demandantes de produtos de varejo? Especificamente daqueles que se vendem em mercados e grandes lojas? Como nesses estabelecimentos são comercializados desde produtos essenciais, de baixa elasticidade-preço (pouco afetados por choques positivos ou negativos de preços) até bens superiores, podemos dizer que os demandantes são, em geral, **toda a população**. Querendo ou não, todos precisamos consumir para sobrevivermos, e os locais "clássicos" (mais seguros do ponto de vista populacional) são as lojas físicas tradicionais.

Desse modo, podemos aproximar nossa demanda utilizando estudos demográficos da região para buscarmos municípios de interesse. Dados que despertam principal interesse são população total, nível médio de renda per capita e níveis de emprego na cidade. A população total é um parâmetro importante pois basicamente nos diz o número de consumidores potenciais total do município. Ela precisa, no entanto, ser ponderada com a medida média de renda e emprego para entendermos o perfil do consumidor e a distribuição da renda. De nada adianta termos muita população na cidade se a renda está concentrada em poucos potenciais demandantes dos produtos de nossos parceiros.

Intuitivamente, essa linha de raciocínio nos leva a pensarmos nas seguintes organizações geográficas, em ordem:

1. Capitais dos estados.
2. Regiões metropolitanas.
3. Polos de alta atividade industrial/tecnológico.

Explicando brevemente, as capitais dos Estados geralmente possuem maior desenvolvimento de infra-estrutura, maiores níveis populacionais e costumam figurar entre os maiores níveis de renda de cada estado. As regiões metropolitanas se aproveitam da conexão com as capitais para logística e da economia de escala que surge da proximidade ao centro urbano. Polos de atividade econômica podem ser geograficamente distante das capitais, mas geralmente empregam alta quantidade de mão-de-obra e também possuem níveis de renda acima da média estadual. O que faremos a seguir é verificar se esses pressupostos se confirmam e escolher cidades possivelmente mais vantajosas para início das operações. Como o propósito é a escolha de 10 cidades, a análise será feita da seguinte maneira: 3 cidades de cada estado, para efeitos de diversificação de riscos mas que possibilitem o aproveitamento do custo operacional dentro de cada estado, mais uma cidade extra em um dos estados, que aparente capacidade de adaptação às atividades da Anthor.

*Obs.: Em uma situação real com maior tempo e dispêndio de esforço disponíveis, minha aproximação seria diferente e mais "objetiva". Se os dados dos municípios não estão disponíveis em formato tabular, eu tabularia ao menos os 20 mais populosos de cada região e criaria algum tipo de métrica para "rankear" as diferentes cidades. Seriam coletados os dados de interesse, normalizados para que não haja problemas de dimensionamento e então ponderados de acordo com critérios definidos humanamente para dar uma nota a cada cidade, de acordo com os critérios estabelecidos. É um dizer comum que o trabalho do cientista de dados está 80% na coleta e tratamento e 20% na análise dos dados, nesse caso não seria diferente.*

## 2.1 Preparando as tabelas com dados temporais

Vamos começar definindo os dados que serão utilizados de base para estimação dos dados de interesse. A partir das informações das cidades com maior maturidade do uso na plataforma, vamos buscar algum indicador que nos forneça insights sobre os primeiros meses de operação em cada cidade. Arbitrariamente, decidi reunir dados sobre os 5 primeiros meses de operação de certo número de cidades. Em certas condições, esses dados se encaixariam na classificação de séries temporais (pois utilizariam-se as datas absolutas como índices para referenciamento dos dados), mas no caso não nos interessa exatamente **quando** a operação começou, sim a evolução através dos meses (mês 1, mês 2, etc.).

Utilizaremos a estrutura de DataFrames da biblioteca pandas para montar essa tabela.

```python
import pandas as pd

colunas = ('Cidade', 'Mês de Operação', 'Demandas Realizadas', 'Workers')

dados = [['Curitiba', 1, 243, 33],
         ['Curitiba', 2, 370, 58],
         ['Curitiba', 3, 461, 75],
         ['Curitiba', 4, 592, 98],
         ['Curitiba', 5, 844, 117],
         ['Ponta Grossa', 1, 9, 4],
         ['Ponta Grossa', 2, 14, 9],
         ['Ponta Grossa', 3, 61, 10],
         ['Ponta Grossa', 4, 75, 9],
         ['Ponta Grossa', 5, 191, 19],
         ['Maringá', 1, 9, 4],
         ['Maringá', 2, 31, 10],
         ['Maringá', 3, 48, 10],
         ['Maringá', 4, 57, 9],
         ['Maringá', 5, 64, 6],
         ['Foz do Iguaçu', 1, 31, 4],
         ['Foz do Iguaçu', 2, 26, 3],
         ['Foz do Iguaçu', 3, 13, 1],
         ['Foz do Iguaçu', 4, 34, 4],
         ['Foz do Iguaçu', 5, 75, 13]]


infos_inicio_operação = pd.DataFrame(dados, columns = colunas)
infos_inicio_operação['Demanda/Worker'] = infos_inicio_operação['Demandas Realizadas']/infos_inicio_operação['Workers']
infos_inicio_operação
```

Com as informações da tabela, realizaremos alguns agrupamentos. Em primeiro momento, parece interessante sabermos as médias por mês de operação e por cidade, para vermos se existe algum tipo de informação que possa ser utilizada.

```python
#agrupando as médias por mês de operação
medias_por_mês = infos_inicio_operação.groupby('Mês de Operação', as_index=False).mean()
display(medias_por_mês)

#agrupando a média por cidade
display(infos_inicio_operação.groupby('Cidade')[['Demandas Realizadas', 'Workers', 'Demanda/Worker']].mean())
```

Desses agrupamentos, vemos algumas coisas que já chamam a atenção: em primeiro lugar, nos primeiros meses a média de demandas por worker é crescente ms com pouca flutuação. Em segundo lugar, o número não flutua muito dentre as diferentes cidades, mesmo com números absolutos de demandas e dimensionamento de operações bastante distintos. Cada worker realizando entre 6 e 8 missões nos primeiros meses, em média, parece algo consistente pelos dados estudados.

## 2.2 Estimando Demanda por regressão

Como exercício prático, vamos tentar estimar a demanda através de uma regressão linear simples. Com a amostra de 20, vamos buscar mais alguns dados das cidades que talvez sejam explicativos e inserí-los na tabela através de joins.
Visto que os dados municipais de número de varejistas estão disponíveis, vamos tentar aproximações através de variáveis proxy. Usaremos os seguintes dados:

- População do município
- IDH do município (proxy de renda e desenvolvimento, não é a melhor mas é a mais disponível no momento da elaboração)
- Mês de operação
- Número de empresas na cidade (como proxy do número de varejistas)

Com os dados que já temos, montaremos um modelo para predição dos dados que buscamos nos novos municípios. Obviamente, uma amostra de 20 gera problema de micronumeridade e consequentemente de fit da regressão: ela ficará "viciada" e se ajustará muito aos dados usados para treino, impactando no teste. No entanto, o raciocício para a resolução do problema em um contexto real é o mesmo, sendo necessário o aumento da amostra de treino e o teste com mais variáveis-chave, conforme se ajusta o modelo. 

Em primeiro lugar, vamos criar uma tabela com as informações dos municípios estudados (extraídas manualmente do IBGE cidades):


```python
header = ('Cidade', 'População', 'IDH', 'Nº Empresas')
data = [['Curitiba', 1948626, 0.823, 104867],
        ['Ponta Grossa', 355336, 0.763, 12579],
        ['Maringá', 430157, 0.808, 26344],
        ['Foz do Iguaçu', 258248, 0.751, 9567],
        ['Salvador', 2886698, 0.759, 57066],
        ['Feira de Santana', 619609, 0.712, 13921],
        ['Lauro de Freitas', 201635, 0.754, 7526],
        ['Fortaleza', 2686612, 0.754, 56922],
        ['São Gonçalo do Amarante', 43890, 0.665, 613],
        ['Maracanaú', 229458, 0.686, 3462],
        ['Recife', 1653461, 0.772, 41701],
        ['Jaboatão dos Guararapes', 706867, 0.717, 7666],
        ['Olinda', 39315, 0.735, 5538],
        ['Porto de Galinhas', 97669, 0.619, 1499]]


infos_cidades = pd.DataFrame(data, columns = header)
infos_cidades
```

Agora vamos inserir os dados na matriz de preditores, ou seja, das observações que tentaremos utilizar para "treinar" o modelo de regressão linear.


```python
info_preditores = pd.merge(infos_inicio_operação, infos_cidades, on='Cidade', how='left')
display(info_preditores)
```

Agora separamos em matrizes de variáveis de features (mês, população e IDH) e target (demandas realizadas). Depois, utilizaremos a classe linear_regression da biblioteca sklearn para "treinar" o modelo com os coeficientes de cada variável explicativa.

```python
X_modelo = info_preditores[['Mês de Operação', 'População', 'IDH', 'Nº Empresas']]

y_modelo = info_preditores[['Demandas Realizadas']]

#importando a regressão linear
from sklearn.linear_model import LinearRegression

#gerando o objeto de regressão
re = LinearRegression()

#treinando com os dados da matriz de regressores
re.fit(X_modelo, y_modelo)

#obtendo os coeficientes de cada regressor e intercepto da regressão
display(re.coef_)
display(re.intercept_)
```

A existência de um intercepto com valor alto de demanda indica fatores que não estão captados na regressão. Em uma situação real, indica a existência de variáveis importantes não incorporadas no modelo (desde que este de fato siga relação linear entre as variáveis). Além disso, dada a amplitude das variáveis usadas como proxy, é extremamente certo que existe algo nível de correlação entre as variáveis explicativas, fator que também afeta a qualidade de uma regressão.
Neste exercício, os fatores de tempo e população atuam positivamente no aumento de de demandas, enquanto que IDH e número de empresas negativamente. Para entender se esses resultados fazem sentido, é necessário buscar explicações teóricas. O fator IDH até pode representar queda de consumo (se o comportamento padrão for de aumento de poupança), mas o número de empresas ser negativo parece ser alguma anomalia. Não há muito a ser feito devido à indisponibilidade de mais detalhamento de informações tanto de dados "externos" quando internos. Mas, novamente, a lógica do "workflow" a ser feito no processo de predição dos dados seria o mesmo, apenas com informações de melhor qualidade e maior amostragem/tratamento de dados. O processo de estimação envolveria tirar e inserir variáveis e avaliar a qualidade dos resultados e das predições, utilizando sets de teste em que pudéssemos evaluar resultados. Com uma amostra pequena de 20, fica inviável criar um set para estes testes. 

Agora, vamos criar outra matriz com os dados a serem preditos pela regressão: os dos municípios indicados. Para mudar um pouco as coisas, criei a matriz em excel e li o csv através do pandas.

```python
matriz_preditos = pd.read_csv('Pasta1.csv', encoding='latin-1', sep=';')

#extraindo algumas linhas para mostrar a forma da tabela
matriz_preditos.sample(10)
```

Faremos agora também o join com a matriz de informações e a separação em variáveis preditoras, para que possamos chamar a regressão linear. A regressão vai criar uma matriz Y com valores preditos a partir dos dados informados e dos parâmetros criados na etapa anterior.

```python
#criando matriz geral
infos_a_predizer = pd.merge(matriz_preditos, infos_cidades, how='left', on='Cidade')
display(infos_a_predizer.sample(10))

#separando a matriz de preditores X
X_predit = infos_a_predizer[['Mês de Operação', 'População', 'IDH', 'Nº Empresas']]
y = re.predict(X_predit)

infos_a_predizer['Demanda Prevista'] = y

display(infos_a_predizer)
```

Podemos ver que há vários municípios em que a regressão nos primeiros meses possui demanda projetada de valores negativos. Isso ocorre pois o fator "Mês de operação" está com grande peso no resultado, carregando os fatores de tempo para aquisição de novos parceiros, geração de demandas, etc. Obviamente, não existe demanda negativa e tal situação provavelmente não ocorreria na vida real. Como a regressão está utilizando de muitas variáveis proxy e é bem frágil, esse tipo de evento ocorre. Mesmo assim, podemos ver que o restante dos dados não parece tão absurdo, visto que as tendências de aumento são respeitadas. O modelo provavelmente teria bom poder preditivo se munido de mais informações e ajustado adequeadamente. 

Vamos continuar utilizando esse modelo, a fins abstrativos, e apenas tratar os dados antes de completar a matriz com todas as informações solicitadas. A primeira coisa a fazer é zerar os valores em que a demanda prevista é negativa, visto que não há possibilidade desse tipo de ocorrência em uma situação real.

```python
#biblioteca matricial, funciona muito bem junto ao Pandas
import numpy as np

infos_a_predizer['Demanda Prevista'] = np.where(infos_a_predizer['Demanda Prevista']<0, 0, 
                                                infos_a_predizer['Demanda Prevista'])

display(infos_a_predizer)
```

Agora, vamos utilizar as médias de demanda realizadas por trabalhor obtidas do agrupamento anterior em meses e por fim apenas dividir o número de demandas pelas médias para termos um resultado de número de trabalhadores estimado, concluindo nossa matriz completa de predições. A matriz completa está à mostra no próximo capítulo.

```python
#juntando as informações médias do agrupamento realizado anteriormente com a tabela de predições
infos_pred_completas = pd.merge(infos_a_predizer, medias_por_mês[['Mês de Operação', 'Demanda/Worker']], on='Mês de Operação', how='left')

#criando uma nova coluna com o número de workers obtido a partir da divisão do total de demandas pela média de realizações.
infos_pred_completas['Workers Estimados'] = round(infos_pred_completas['Demanda Prevista']/infos_pred_completas['Demanda/Worker'])
```
